DATA CLEANING

```python
import pandas as pd
vl= pd.read_csv("laptopData.csv") 
```

```python
cat = ['Company', 'TypeName', 'Cpu', 'Gpu', 'OpSys', 'ScreenResolution', 'Memory']

for col in cat:
    vl[col] = vl[col].fillna(vl[col].mode()[0]) 
```

```python
vl = vl.drop("Unnamed: 0", axis=1) 
```

```python
vl 
```

```python

vl['Price'] = vl.groupby('Company')['Price'].transform(lambda x: x.fillna(x.median())) 
```

```python
vl 
```

```python
vl.dtypes 
```

```python
vl["Ram"] = vl["Ram"].ffill().bfill() 
```

```python
vl["Ram"]=vl["Ram"].str.extract(r'(\d)') 
```

```python
vl["Weight"]=vl["Weight"].str .extract(r'(\d)') 
```

```python
vl.dtypes 
```

```python
vl.rename(columns={'Weight':'Weight(KG)'},inplace=True)
```

```python
vl['Inches']=pd.to_numeric(vl['Inches'],errors='coerce') 
```

```python
vl['Weight(KG)']=pd.to_numeric(vl['Weight(KG)'],errors='coerce') 

```

```python
vl['Ram']=pd.to_numeric(vl['Ram'],errors='coerce') 
```

```python
vl['Ram']=pd.to_numeric(vl['Ram'],errors='coerce') 
```

```python
vl.dtypes 
```

```python
vl.isnull().sum()
```

```python
mean_inches = vl['Inches'].mean()
print(mean_inches)
vl['Inches']=vl['Inches'].fillna(mean_inches)
```

```python
vl
```

```python
mean_inches = vl['Weight(KG)'].mean()
print(mean_inches)
vl['Weight(KG)']=vl['Weight(KG)'].fillna(mean_inches)
```

```python
vl.rename(columns={'Ram':'Ram'},inplace=True)
```

```python
vl['Weight(KG)']=pd.to_numeric(vl['Weight(KG)'],errors='coerce') 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.boxplot(vl[['Ram', 'Weight(KG)','Inches']])
plt.xticks([1,2,3,],['Ram', 'Weight(KG)','Inches'])
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
q3 = vl['Inches'].quantile(0.75)
q1 = vl['Inches'].quantile(0.25)
iqr = q3 - q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr

med_val=vl['Inches'].median()

vl.loc[(vl['Inches']<lower_bound) | (vl['Inches']>upper_bound),'Inches']=round(med_val,0)
```

```python
q3 = vl['Weight(KG)'].quantile(0.75)
q1 = vl['Weight(KG)'].quantile(0.25)
iqr = q3 - q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr

med_val=vl['Weight(KG)'].median()

vl.loc[(vl['Weight(KG)']<lower_bound) | (vl['Weight(KG)']>upper_bound),'Weight(KG)']=round(med_val,0)
```

```python
plt.boxplot(vl[['Ram', 'Weight(KG)','Inches']])
plt.xticks([1,2,3,],['Ram', 'Weight(KG)','Inches'])
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    x='Ram',
    y='Price',
    data=vl,
    palette='Set2'  
)
plt.title('Price vs. RAM', fontsize=14)
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.show()

```

```python
# Make sure Cpu_Brand exists
vl['Cpu_Brand'] = vl['Cpu'].apply(lambda x: x.split()[0])

plt.figure(figsize=(12, 6))
sns.boxplot(
    x='Cpu_Brand',
    y='Price',
    data=vl,
    palette='husl'  # Husl is bright & unique
)
plt.title('Price vs. CPU Brand', fontsize=14)
plt.xlabel('CPU Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.stripplot(
    x='Ram',
    y='Weight(KG)',
    data=vl,
    hue='Ram',           
    jitter=True,
    palette='Set2',     
    dodge=False          
)
plt.title('Weight by RAM Size (Stripplot)', fontsize=14)
plt.xlabel('RAM(GB)')
plt.ylabel('Weight(KG)')
plt.legend(title='RAM Size', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.swarmplot(
    x='Gpu_Brand',
    y='Price',
    data=vl,
    palette='Set2'
)

plt.title("Price by GPU Brand (Swarm Plot)")
plt.xticks(rotation=45)
plt.show()
```
